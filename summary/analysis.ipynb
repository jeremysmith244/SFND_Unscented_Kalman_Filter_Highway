```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('output.txt', header=None)
```

```python
df.drop(columns=[0,4],inplace=True)
```

```python
df
```

```python
radar = df.loc[list(range(0,300,2))]
laser = df.loc[list(range(1,300,2))]

print(radar[1].mean())
print(laser[1].mean())
```

```python
plt.plot(df[1]);
plt.plot(df[2]);
plt.plot(df[3]);
```

```python
plt.plot(radar[1]);
plt.plot(radar[2]);
plt.plot(radar[3]);
plt.ylim((0,15));
plt.xlabel('k');
plt.ylabel('NIS Radar');
plt.title('{}% of NIS Above 7.8'.format((radar > 7.8).sum().sum()/(300*3)*100))
plt.legend(['Car1', 'Car2', 'Car3'])
plt.axhline(7.8,color='r');
plt.savefig('Radar_NIS.png');
```

```python
plt.plot(laser[1]);
plt.plot(laser[2]);
plt.plot(laser[3]);
plt.ylim((0,15));
plt.xlabel('k');
plt.ylabel('NIS Laser');
plt.title('{}% of NIS Above 5.9'.format(str((laser > 5.9).sum().sum()/(300*3)*100)[:3]))
plt.legend(['Car1', 'Car2', 'Car3'])
plt.axhline(5.9,color='r');
plt.savefig('Laser_NIS.png');
```
